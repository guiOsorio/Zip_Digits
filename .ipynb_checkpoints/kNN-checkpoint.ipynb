## KNN ALGORITHM

```python
class KNN:
    
    '''
    
    KNN algorithm implementation using PCA as the dimensionality reduction technique, and euclidean distance as
    the distance between points.
    
    Inputted datasets need to be divided into features and labels, and need to be pandas dataframes.
    
    '''
    
    def __init__(self, train_features, train_labels, test_features, test_labels):
        self.train_features = train_features
        self.train_labels = train_labels
        self.test_features = test_features
        self.test_labels = test_labels.to_numpy()
        
        # Perform dimensionality reduction with PCA
        self.principals_train = self.PCA(self.train_features)
        self.principals_test = self.PCA(self.test_features)
        
        # Store predictions for accuracy calculation
        self.preds = None
        
        
    def PCA(self, features):
        # For more info on this implementation, check https://guiosorio.github.io/Gui-DS-Learning/2022/07/13/PCA.html
        
        col_means = np.array(np.mean(features, axis=0))
        centered_M = features - col_means
        covar = np.cov(centered_M.T)
        eig_vals, eig_vecs = np.linalg.eig(covar)
        principals = eig_vecs.T.dot(centered_M.T).T
        
        return principals
    
    
    def PCA2_viz(self):
        ''' Visualize first 2 principal components of the training set '''
        
#         self.principals_train = self.PCA(self.train_features)
        
        # Take first 2 principal components
        principals2 = self.principals_train[:, 0:2]
        # Plot 2 principal components and color by the respective label
        # (converted training labels df to numpy array to fix minor bug in labels)
        sns.scatterplot(x=principals2[:,0], y=principals2[:,1], hue=self.train_labels.to_numpy())
        plt.show()
        
        
    def find_all_nns(self, k, pr_comp):
        ''' Find the nearest neighbors of each point in the test set using euclidean distance '''
        
        # List to store the nearest neighbors of each point in the test set
        k_nearest_neighbors = [] 
        
        # Take only the specified number of principal components from principals of features sets (default is 2)
        pr_comps_test = self.principals_test[:, 0:pr_comp]
        pr_comps_train = self.principals_train[:, 0:pr_comp]
        
        # Calculate distance between each point in the test set to each point in the training set
        for p in pr_comps_test:
            dists = np.array([])

            for pp in pr_comps_train:
                # Euclidean distance
                dist = np.linalg.norm(p - pp)
                dists = np.append(dists, dist)
            
            # np.argpartition - returns a list with the k smallest element's indexes in an np array
            # (np array, k smallest elements)
            idx = np.argpartition(dists, k)
            idx_kneighbors = idx[:k] # np array with k nearest neighbors of p (as index in training features)
            
            k_nearest_neighbors.append(idx_kneighbors)
            
        return k_nearest_neighbors
            
        
    def predict(self, k, pr_comp=2):
        '''
        
            -> k is the hyperparameter of number of nearest neighbors
            -> pr_comp represents the number of principal components we intend to use in the euclidean distance 
                calculation (defaults to 2)
                
        '''
        
        # Calculate distance between each test data point and training data points using euclidean distance.
        # Keep k closest neighbors.
        # Each element of the list 'nns' represents the k nearest neighbors 
        ## of the data point of the same index in the test set
        all_nns = self.find_all_nns(k, pr_comp)

        # Map nearest neighbors to each respective label in training set
        all_nns_labels = [] # store mapped labels
        for nns in all_nns:
            curr_labels = [] # store labels of nearest neighbors of current element
            for nn in nns:
                label = self.train_labels.loc[nn]
                curr_labels.append(label)
            all_nns_labels.append(curr_labels)       
        
        # 'nns_labels' is now a list of lists, each with the label for the k nearest neighbors of a point in the test set
        # Predict based on k nearest neighbors - most frequent label (mode)
        preds = []
        for nns_labels in all_nns_labels:
            pred = mode(nns_labels)
            preds.append(pred)

        self.preds = preds
        return preds
    
    
    def accuracy(self):
        ''' Calculate accuracy of the model using the test set '''
        
        correct = 0
        total = 0
        for i in range(len(self.preds)):
            if self.preds[i] == self.test_labels[i]:
                correct += 1
            total += 1
        
        acc = correct / total
        return acc
            
```

## KNN CLASSIFICATION OF 2s AND 3s

```python
# Initialize kNN instance
digits23_KNN = KNN(df_train_features23, df_train_labels23, df_test_features23, df_test_labels23)

# Visualize reduced training set
digits23_KNN.PCA2_viz()
```

```python
# Predict with k = 3
preds23 = digits23_KNN.predict(3)

# Calculate accuracy
digits23_KNN.accuracy()
```

```python
# Plot different Ks with corresponding accuracy to choose the optimal hyperparameter
accuracies23 = {}

# Only try out odd Ks so there are no tie issues
for i in range(1, 20, 2):
    test_KNN = KNN(df_train_features23, df_train_labels23, df_test_features23, df_test_labels23)
    test_preds23 = test_KNN.predict(i)
    acc = test_KNN.accuracy()
    accuracies23[i] = acc
```

```python
plot_23 = sns.scatterplot(x=accuracies23.keys(), y=accuracies23.values())
plt.xlabel('K')
plt.ylabel('Accuracy')
plot_23.set_xticks(range(1,20,2))
```

```python
# 3 seems to be the ideal choice for hyperparameter k
print('Accuracy with optimal hyperparameter (k=3): ', round(accuracies23[3], 4))
```
